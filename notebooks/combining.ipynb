```python
import pandas as pd
```

```python
df1 = pd.read_csv("../data/interim/car_sales_data.csv")
print(df1.shape)
df1.head(3)
```

```python
df2 = pd.read_csv("../data/interim/autoscout24-germany-dataset.csv")
print(df2.shape)
df2.head(3)
```

### Make Model Year match

```python
df3 = pd.merge(df1, df2[['make', 'model', 'year', 'fuel', 'gear', 'hp', 'mileage', 'sale_price_used']], on=['make', 'model', 'year'], how='left')
print(df3.shape)
df3.head(30)
```

```python
df3 = df3[['date_sold_new', 'make', 'model', 'year', 'fuel', 'gear', 'hp', 'sale_price_new', 'car_age']]
```

```python
df3 = df3.dropna()
```

```python
df3 = df3.drop_duplicates()
```

```python
df3
```

```python
df3 = df3.rename(columns={"car_age": "car_age_sold_new", "mileage": "used_mileage"})
```

```python
df3['price_difference_sold_used'] = df3['sale_price_used']- df3['sale_price_new']
```

```python
df3.head(3)
```

```python
df3.to_csv("../data/clean/data.csv", index=False)
```
