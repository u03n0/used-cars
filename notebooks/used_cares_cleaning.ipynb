```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/autoscout24-germany-dataset.csv")
print(df.shape)
df.head(10)
```

```python
df.info()
```

```python
cols = ['make', 'model', 'fuel', 'gear', 'offerType']
for col in cols:
    df[col] = df[col].str.lower()
```

```python
df = df.drop_duplicates()
df = df.drop(columns=["offerType"])
```

```python
df = df.rename(columns={"price": "sale_price_used"})
```

```python
df.to_csv("../data/interim/autoscout24-germany-dataset.csv", index=False)
```
