```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/car_sales_data.csv")
print(df.shape)
df.head(10)
```

## lower column names, and fix spacing.

```python
def fix_column_names(_list):
    return [elem.replace(" ", "_").lower() for elem in _list]

df.columns = fix_column_names(df.columns)
```

```python
df.head(3)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df['car_age'] = df['date'].dt.year - df['car_year']
```

```python
df = df.drop(columns=['commission_rate', 'commission_earned', 'salesperson', 'customer_name'])
```

```python
df = df.rename(columns={'car_make':'make', "car_model": "model", "car_year": "year", "sale_price": "sale_price_new", "date": "date_sold_new"})
```

```python
cols = ['make', 'model']
for col in cols:
    df[col] = df[col].str.lower()
```

```python
df.head(3)
```

```python
df.to_csv("../data/interim/car_sales_data.csv", index=False)
```
